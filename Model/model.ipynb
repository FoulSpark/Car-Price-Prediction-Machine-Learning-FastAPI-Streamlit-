Car Price prdiction Model

```python
import pandas as pd
import numpy as np
from pathlib import Path
import joblib
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
data = pd.read_csv("cardekho_data (1).csv")
data = data.drop(columns=["Car_Name"])
x = data.drop(columns=["Selling_Price"])
y = data["Selling_Price"]
print(x)
```

```python
print(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_columns = ["Fuel_Type","Seller_Type","Transmission"]

ct = ColumnTransformer(
    transformers=[
        ("encoder", OneHotEncoder(), categorical_columns)
    ],
    remainder="passthrough"
)

x = ct.fit_transform(data)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train ,y_test = train_test_split(x,y,random_state=42,test_size=.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100,criterion="squared_error")
model.fit(x_train,y_train)
```

```python
yprd = model.predict(x_test)
print(np.concatenate((y_test.to_numpy().reshape(len(y_test),1) , yprd.reshape(len(yprd),1)
                    ),1))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print("MAE:", mean_absolute_error(y_test, yprd))
print("MSE:", mean_squared_error(y_test, yprd))
print("RMSE:", np.sqrt(mean_squared_error(y_test, yprd)))
print("R2 Score:", r2_score(y_test, yprd))
```
